```python
popular_songs=spotify_data.query("popularity>90",inplace=False).sort_values("popularity",ascending=False).head(10)
popular_songs
```

```python
spotify_data.set_index("release_date",inplace=True)
spotify_data.index=pd.to_datetime(spotify_data.index)
spotify_data.head()
```

```python
spotify_data[["artists"]].iloc[18]
```

```python
spotify_data["duration"]=spotify_data["duration_ms"].apply(lambda x:round(x/1000))
spotify_data.drop("duration_ms",inplace=True, axis=1)
```

```python
spotify_data.duration.head()
```

###pearson corelation bond between 2 variables

```python
spotify_data_corelation=spotify_data.drop(["explicit","key","mode"],axis=1).corr(method="pearson")
plt.figure(figsize=(15,10))
heatmap=sns.heatmap(spotify_data_corelation,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="red")
heatmap.set_title("Corelation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_spotify_data=spotify_data.sample(int(0.004*len(spotify_data)))
```

```python
sample_spotify_data.head()
```

```python
plt.figure(figsize=(14,5))
sns.regplot(data=sample_spotify_data,y="loudness",x="energy",color="black").set(title="Loudness vs Energy correlation ")
```

```python
plt.figure(figsize=(14,5))
sns.regplot(data=sample_spotify_data,y="popularity",x="acousticness",color="green").set(title="Populairty vs Acousitcness correlation ")
```

```python
spotify_data["dates"]=spotify_data.index.get_level_values("release_date")
spotify_data.dates=pd.to_datetime(spotify_data.dates)
years=spotify_data.dates.dt.year
                                                            
```

```python
#pip install --user seaborn==0.11.0
```

```python
sns.displot(years,discrete=True, aspect=3,height=5,kind="hist").set(title="Number of songs per year");
```

```python
total_duration=spotify_data.duration
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_duration,ax=ax,errwidth=False).set(title="Years vs duration");
plt.xticks(rotation=90)
```
