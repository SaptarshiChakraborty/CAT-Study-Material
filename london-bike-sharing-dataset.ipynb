```python
##changing column name 
new_cols_dict={
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_c',
    't2':'temp_feels_like_c',
    'hum': 'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}
london_df.rename(new_cols_dict,axis=1,inplace=True)
```

```python
london_df.head(5)
```

```python
##changing humidity values to percentage 

london_df.humidity_percent=london_df.humidity_percent/100
```

```python
london_df.head(5)
```

```python
##creating season dictionary to map values according to the numbers
season_dict={
    '0.0':'spring',
    '1.0': 'summer',
    '2.0': 'autumn',
    '3.0': 'winter'
}
##creating weather dictionary 
weather_dict={
    '0.0':'clear',
    '1.0':'scattered clouds',
    '2.0':'broken cloudsr',
    '3.0':'cloudy',
    '4.0':'rainy',
    '5.0':'rainy with thunderstorm',
    '6.0':'Snowy'
}
```

```python
london_df.head(5)
```

```python
##changing the season column to string
london_df.season=london_df.season.astype('str')

##mapping the season values ranging from0-3
london_df.season=london_df.season.map(season_dict)

#changing weather value to string
london_df.weather=london_df.weather.astype("str")
#mapping the weather value accordingly to the range
london_df.weather=london_df.weather.map(weather_dict)
```
